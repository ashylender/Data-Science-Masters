Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.


```python
import plotly.graph_objects as go
import seaborn as sns
```

```python
ds = sns.load_dataset('titanic')
ds
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(x = ds.age, y = ds.fare, mode = "markers"))
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px
```

```python
data = px.data.tips()
```

```python
data
```

```python
fig = px.bar(data, x='sex', y='smoker')
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px
```

```python
fig = px.histogram(data, x = 'sex', y = 'total_bill',color = 'day', pattern_shape ='smoker')
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
import plotly.express as px
ds = px.data.iris()
```

```python
ds
```

```python
fig = px.scatter_matrix(ds, color='species')
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

The displot figure displays a combination of statistical representation of numerical data, such as Hitsogram, Kernel density estimation or normal curve and rug plot

```python
## To plot displot we have to import figure_factory module of plotly.
import plotly.figure_factory as ff
```

```python
from scipy.special import xlogy

import numpy as np
x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot']
fig = ff.create_distplot(hist_data,group_labels)
fig.show()
```
